````python
import mysql.connector
import pandas as pd
````

````python
mydb = mysql.connector.connect(
    host = 'localhost',
    user = 'root',
    passwd = '1411',
    database = 'sakila'
)
````

````python
mycursor = mydb.cursor()

# create function
def sql_df(yourQuery):
    mycursor.execute(yourQuery)
    myresult = mycursor.fetchall()
    df = pd.DataFrame(myresult, columns = mycursor.column_names)
    return df
````

***1. Tampilkan daftar ```10 film komedi dengan durasi tersingkat.``` Urutkan data berdasarkan film dengan durasi terpendek. Kolom yang diwajibkan tampil adalah ```title, category dan length.``` Output yang diharapkan:***

````python
# Menampilkan kolom title dari tabel film, name dari tabel category, dan length dari tabel film
# menggabungkan tabel film dengan film_category melalui film_id lalu digabungkan dengan tabel category melalui category_id
# dengan selection tabel name = comedy
# di urutkan dari dari terendah ke teringgi melalui length
# di tampilkan 10 data teratas

sql_df('''
SELECT film.title, category.name, film.length
FROM film  JOIN film_category USING (film_id) JOIN category USING (category_id)
WHERE category.name = 'comedy'
ORDER BY film.length
LIMIT 10
''')
````

***2. Tampilkan daftar lengkap ```kategori film beserta jumlah film tiap kategori & rata-rata harga sewa DVD film tiap kategori.``` Urutkan data dari kategori dengan jumlah film terbanyak. Kolom yang diwajibkan ada minimal adalah ```kategori, jumlah film dan rata-rata harga sewa.``` Output yang diharapkan:***

````python
# Menampilkan kolom name dari tabel category sebagai Kategori, menghitung jumlah title dari tabel film sebagai JumlahMovie,
# dan menghitung rata2 dari rental_rate tabel film sebagai RataHargaSewa
# menggabungkan tabel film dengan film_category melalui film_id lalu digabungkan dengan tabel category melalui category_id
# di group by dari kolom Kategory
# di urutkan dari dari tertingi ke terendah melalui JumlahMovie

sql_df('''
SELECT category.name AS Kategori, count(film.title) as JumlahMovie, avg(film.rental_rate) as RataHargaSewa
FROM film JOIN film_category USING (film_id) JOIN category USING (category_id)
GROUP BY Kategori
ORDER BY JumlahMOvie DESC
''')
````

***3. Tampilkan daftar ```10 aktor/aktris yang paling banyak membintangi film.``` Kolom yang ditampilkan minimal: ```id aktor, nama depan, nama belakang dan jumlah film yang dibintangi``` kemudian urutkan dari aktor/aktris yang membintangi film terbanyak. Output yang diharapkan:***

````python
# Menampilkan kolom actor_id dari tabel actor, first_name dari tabel actor, last_name dari tabel actor,
# dan menghitung jumlah data dari tittle tabel film sebagai jumlah_movie
# menggabungkan tabel actor dengan film_actor melalui actor_id lalu digabungkan dengan tabel film melalui film_id
# di group by dari kolom actor_id dari tabel actor
# di urutkan dari dari tertingi ke terendah melalui jumlah_movie, actor.actor_id
# dibatasi 10 data teratas

sql_df('''
SELECT actor.actor_id, actor.first_name, actor.last_name, count(film.title) AS jumlah_movie
FROM actor JOIN film_actor USING (actor_id) JOIN film USING (film_id)
GROUP BY actor.actor_id
ORDER BY jumlah_movie DESC, actor.actor_id DESC
LIMIT 10
''')
````

***4. Dari soal sebelumnya diketahui ```Gina Degeneres``` merupakan aktris yang paling banyak membintangi film, dengan total 42 judul film. Kategori film apakah yang paling banyak dibintanginya? Untuk mengetahuinya, tampilkan daftar ```kategori film beserta jumlah film``` yang pernah dibintangi oleh ```Gina Degeneres.``` Kolom yang diwajibkan ada yaitu ```kategori film dan jumlah film yang dibintangi.``` Output yang diharapkan:***

````python
# Menampilkan kolom name dari tabel category, first_name dari tabel actor,
# dan menghitung jumlah data dari name pada tabel category sebagai jumlah_movie
# menggabungkan tabel actor dengan film_actor melalui actor_id lalu digabungkan dengan tabel film melalui film_id
# lalu digabungkan dengan tabel film_category melalui film_id lalu digabungkan dengan tabel category melalui category_id
# dilakukan selection melalui actor.actor_id = 107
# di group by dari kolom name dari category

sql_df('''
SELECT category.name, count(category.name) AS jumlah_Movie
FROM actor JOIN film_actor USING (actor_id) JOIN film USING (film_id) 
JOIN film_category USING (film_id) JOIN category USING (category_id)
WHERE actor.actor_id = 107
GROUP BY category.name
''')
````

***5. Dari soal sebelumnya diketahui ```Gina Degeneres``` paling banyak membintangi film bergenre science-fiction, dengan total ```7``` judul film. Tampilkan daftar ```judul film sci-fi``` yang pernah dibintangi oleh ```Gina Degeneres.``` Kolom yang diwajibkan ada yaitu ```judul film dan kategorinya.``` Output yang diharapkan:***

````python
# Menampilkan kolom tittle dari tabel film, name dari tabel category sebagai category
# menggabungkan tabel actor dengan film_actor melalui actor_id lalu digabungkan dengan tabel film melalui film_id
# lalu digabungkan dengan tabel film_category melalui film_id lalu digabungkan dengan tabel category melalui category_id
# dilakukan selection melalui actor.actor_id = 107 dan category.name = 'Sci-Fi'


sql_df('''
SELECT film.title, category.name AS category
FROM actor JOIN film_actor USING (actor_id) JOIN film USING (film_id) 
JOIN film_category USING (film_id) JOIN category USING (category_id)
WHERE actor.actor_id = 107 AND category.name = 'Sci-Fi'
''')
````
